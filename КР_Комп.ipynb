```python
# Подключение библиотек
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Показ данных внутри файла
df = pd.read_csv("winter.csv", sep=",")
df
```

Анализ данных

```python
# Определение количества стран в датафрейме
kol_count = df['Country'].nunique()
print(f"Количество стран: {kol_count}")
```

```python
# Определение количества спортсменов по странам
SporMen_Count = df.groupby('Country')['Athlete'].nunique()
print(f"Количество спортсменов по странам: {SporMen_Count}")
```

```python
# Определение количества спортсменов по годам в разрезе стран
SporMen_Count_Year = df.groupby(['Year', 'Country'])['Athlete'].nunique()
print(f"Количество спортсменов по годам в разрезе стран: {SporMen_Count_Year}")
```

```python
# Определение количества спортсменов по видам спорта и странам
SporMen_Count_Sport = df.groupby(['Sport', 'Country'])['Athlete'].nunique()
print(f"Количество спортсменов по видам спорта и странам: {SporMen_Count_Sport}")
```

```python
# Определение количества видов спорта по годам
Kol_Sport_Year = df.groupby('Year')['Sport'].nunique()
print(f"Количество видов спорта по годам: {Kol_Sport_Year}")
```

```python
# Определение количества участников по половому признаку в разрезе видов спорта
SportMen_GenderPaty = df.groupby(['Sport', 'Gender'])['Athlete'].nunique()
print(f"Количество участников по половому признаку в разрезе видов спорта: {SportMen_GenderPaty}")
```

```python
# Определение количества медалей по годам и странам
Kol_Medals_Year_Count = df.groupby(['Year', 'Country'])['Medal'].count()
print(f"Количество медалей по годам и странам: {Kol_Medals_Year_Count}")
```

```python
# Определение количества медалей по годам и полу
Kol_Medals_Year_GenderParty = df.groupby(['Year', 'Gender'])['Medal'].count()
print(f"Количество медалей по годам и полу: {Kol_Medals_Year_GenderParty}")
```

```python
# Количество полученных медалей отлетом за года
Kol_Medals_SportMen_Year = df.groupby(['Year', 'Athlete'])['Medal'].count().reset_index()
Kol_Medals_SportMen_Year = Kol_Medals_SportMen_Year.loc[Kol_Medals_SportMen_Year.groupby('Year')['Medal'].idxmax()]

print(f"Самые маститые спортсмены по годам:\n {Kol_Medals_SportMen_Year}")
```

```python
# Определение стран, участвовавших каждый год
Count_Year = df.groupby('Country')['Year'].unique()
Count_Year = Count_Year[Count_Year.apply(lambda x: set(x) == set(df['Year'].unique()))]

print(f"Страны, участвовавшие каждый год: {Count_Year}")
```

```python
# Определение спортсменов, участвовавших больше всего и меньше всего
Max_SportMan = df.groupby('Athlete')['Athlete'].count().idxmax()
Min_SportMan = df.groupby('Athlete')['Athlete'].count().idxmin()
print(f"Спортсмен, учавствовавший больше всего: {Max_SportMan}")
print(f"Спортсмен, учавствовавший меньше всего: {Min_SportMan}")
```

Визуализация данных

```python
# Спортсмены по странам
SportMan_Country = df.groupby('Country')['Athlete'].nunique()

# Сортируем значения для более наглядного графика
SportMan_FromCountry = SportMan_Country.sort_values(ascending=False)

# Построение графика
plt.figure(figsize=(10, 6))
SportMan_FromCountry.plot(kind='bar', color='green')
plt.title('Количество спортсменов по странам')
plt.xlabel('Страна')
plt.ylabel('Количество спортсменов')
plt.xticks(rotation=45, ha='right')  # Поворот подписей оси x для лучшей читаемости
plt.tight_layout()

# Показываем график
plt.show()
```

```python
# Построение ящиков с усами для индекса "Количество медалей" по каждой стране
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Country', y='Medal', palette='viridis')
plt.title('Распределение количества медалей по странам')
plt.xlabel('Страна')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45, ha='right')  # Поворот подписей оси x для лучшей читаемости
plt.tight_layout()

# Показываем график
plt.show()
```

```python
# Фильтруем только золотые медали
gold_medals_df = df[df['Medal'] == 'Gold']

# Построение диаграммы плотности распределения золотых медалей по странам
plt.figure(figsize=(12, 8))
sns.countplot(data=gold_medals_df, x='Country', palette='viridis')
plt.title('Распределение золотых медалей по странам')
plt.xlabel('Страна')
plt.ylabel('Количество золотых медалей')
plt.xticks(rotation=45, ha='right')  # Поворот подписей оси x для лучшей читаемости
plt.tight_layout()

# Показываем график
plt.show()
```

```python
# Построение диаграммы плотности распределения количества медалей по полу
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='Medal', palette='viridis')
plt.title('Распределение количества медалей по полу')
plt.xlabel('Пол')
plt.ylabel('Количество медалей')
plt.tight_layout()

# Показываем график
plt.show()
```

```python
# Построение столбчатой диаграммы количества медалей по странам
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Country', hue='Medal', palette='viridis')
plt.title('Распределение количества медалей по странам')
plt.xlabel('Страна')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45, ha='right')  # Поворот подписей оси x для лучшей читаемости
plt.tight_layout()

# Показываем график
plt.show()
```
