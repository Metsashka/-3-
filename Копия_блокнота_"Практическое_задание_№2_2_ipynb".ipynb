```python
# Подключение библиотеки

import pandas as pd
import numpy as np
```

```python
# Задание 1

# Загрузка данных из txt файла в DataFrame
df = pd.read_csv('wr88125.txt', delimiter=';')  # Укажите путь к вашему txt файлу и разделитель данных

# Добавляем заголовки к столбцам
#df.columns = ['индекс ВМО','год','месяц','день','минимальная температура воздуха','средняя температура воздуха','максимальная температура воздуха','количество осадков']
df.columns= ['index', 'year', 'month', 'day', 'min_t', 'average_t', 'max_t', 'rainfall']

# Отображаем DataFrame с новыми заголовками
print(df)
```

```python
# Задание 2

# Предположим, что df - ваш DataFrame
df.drop(columns=['index'], inplace=True)

# Вывод DataFrame после удаления столбца 'index'
print(df)
```

```python
# Задание 3-4-5

# Проверка строк с пропущенными значениями
missing_values_in_rows = df.isnull().sum(axis=1)
rows_with_missing_values = df[missing_values_in_rows > 0]

# Отображение DataFrame информации и строк с пропущенными значениями
print("Информация о DataFrame:")
print(df.info())

print("\nСтроки данных с пропущенными значениями:")
print(rows_with_missing_values)
```

```python
# Задание 6

# Подсчет пропущенных значений по каждому году
missing_values_by_year = df['year'].isnull().groupby(df['year']).sum().sort_values(ascending=False)

# Находим год с наибольшим количеством пропусков
year_with_most_missing_values = missing_values_by_year.idxmax()

print(f"В данных за {year_with_most_missing_values} больше всего пропусков.")
print("Количество пропущенных значений по каждому году:")
print(missing_values_by_year)
```

```python
# Задание 7

# Объединяем столбцы 'year', 'month', 'day' в столбец 'Дата' в формате 'yyyy-mm-dd' и преобразуем в формат datetime
df['Дата'] = pd.to_datetime(df[['year', 'month', 'day']].astype(str).apply(lambda x: '-'.join(x), axis=1))

# Выводим DataFrame с новым столбцом 'Дата' в формате datetime
print(df)
```

```python
# Задание 8
# Предварительно добавьте столбцы 'temperature_range' и 'days_without_rain' с начальными значениями 0

for index, row in df.iterrows():
    # Проверка числовых значений перед преобразованием
    try:
        max_temp = float(row['max_t'].strip())
        min_temp = float(row['min_t'].strip())
    except ValueError:  # Обработка ошибки конвертации
        max_temp, min_temp = 0, 0  # Можно использовать другие значения по умолчанию

    # Расчет размаха температур
    temperature_range = max_temp - min_temp

    # Проверка количества дней без осадков
    days_without_rain = 0

    current_index = index
    while current_index >= 0 and df.loc[current_index, 'rainfall'] == 0:
        days_without_rain += 1
        current_index -= 1

    # Обновление значений в DataFrame
    df.at[index, 'temperature_range'] = temperature_range
    df.at[index, 'days_without_rain'] = days_without_rain

# Вывод обновленного DataFrame
print(df)
```

```python
# Задание 9
# Предположим, что df - ваш DataFrame
max_drought_period = 0
current_drought_period = 0

for index, row in df.iterrows():
    if row['rainfall'] == 0:
        current_drought_period += 1
    else:
        if current_drought_period > max_drought_period:
            max_drought_period = current_drought_period
        current_drought_period = 0

# Проверяем для последней последовательности, если она является самой длинной
if current_drought_period > max_drought_period:
    max_drought_period = current_drought_period

print(f"Самый длинный период засухи составляет {max_drought_period} дней.")
```

```python
# Задание 10

# Проверка данных перед преобразованием
df['average_t'] = pd.to_numeric(df['average_t'], errors='coerce')  # Преобразование в числовой формат, значения, которые не могут быть сконвертированы, будут заменены на NaN
df['rainfall'] = pd.to_numeric(df['rainfall'], errors='coerce')  # Аналогично для столбца 'rainfall'

# Вычисление среднегодовой температуры и общего количества осадков для каждого года
yearly_data = df.groupby('year').agg({'average_t': 'mean', 'rainfall': 'sum'})

# Создание объектов Series для среднегодовой температуры и общего количества осадков
average_temperature_series = pd.Series(yearly_data['average_t'], name='average_temperature')
total_rainfall_series = pd.Series(yearly_data['rainfall'], name='total_rainfall')

# Вывод результатов
print("Среднегодовая температура:")
print(average_temperature_series)

print("\nОбщее количество осадков:")
print(total_rainfall_series)
```

```python
# Задание 11

# Находим самый теплый и самый холодный год
hottest_year = df.loc[df['average_t'].idxmax()]['year']
coldest_year = df.loc[df['average_t'].idxmin()]['year']

print(f"Самый теплый год: {hottest_year}")
print(f"Самый холодный год: {coldest_year}")
```

```python
# Задание 12

import matplotlib.pyplot as plt
# Находим год с максимальным количеством осадков
max_rainfall_year = df['average_t'].idxmax()
max_rainfall_amount = df['average_t'].max()

# Находим год с минимальным количеством осадков
min_rainfall_year = df['average_t'].idxmin()
min_rainfall_amount = df['average_t'].min()

print(f"Год с наибольшим количеством осадков: {max_rainfall_year} ({max_rainfall_amount} мм/год)")
print(f"Год с наименьшим количеством осадков: {min_rainfall_year} ({min_rainfall_amount} мм/год)")

# Построение графика изменения температуры
average_temperature_series.plot()
plt.xlabel('Год')
plt.ylabel('Температура')
plt.title('Изменение температуры по годам')
plt.show()

# Построение столбиковой диаграммы для количества осадков по годам
total_rainfall_series.plot.bar()
plt.xlabel('Год')
plt.ylabel('Количество осадков, мм')
plt.title('Количество осадков в каждом году')
plt.show()
```

```python
# Задание 13-14

# Установка порога температуры
threshold_temperature = -30  # Можно использовать другие значения, такие как -10, -35, -40

# Нахождение лет, когда средняя температура была ниже установленного порога
years_below_threshold = df[df['average_t'] < threshold_temperature]['year'].unique()

num_years_below_threshold = len(years_below_threshold)

print(f"Количество лет, когда средняя температура была ниже {threshold_temperature}°C: {num_years_below_threshold}")
print(f"Годы, в которых наблюдалась средняя температура ниже {threshold_temperature}°C: {years_below_threshold}")
```

```python
# Задание 13-15
# Условия для выбора нужных лет
above_temp_condition = df['average_t'] > 27
rainfall_condition = df['rainfall'] > 3

# Выбор лет, когда оба условия выполняются
selected_years = df[above_temp_condition & rainfall_condition]['year'].unique()

num_selected_years = len(selected_years)

print(f"Количество лет, когда средняя температура была выше 27°C и количество дней без осадков было больше 3: {num_selected_years}")
print(f"Годы, удовлетворяющие этим условиям: {selected_years}")
```
