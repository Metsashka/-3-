#  Практическое задание №3.1. Визуализация с помощью библиотеки Matplotlib

Самостоятельно построить рисунок, с основными элементами изображения:
с линиями оси координат, подписями осей, подписями основных и промежуточных делений, линиями графика (линии графика могут быть любые!),
маркерами, легендой, заголовком.
                                                                                                      
Пример рисунка

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator, FuncFormatter

def hyperbola(x, a, b, h, k, upper_limit, lower_limit):
    y = k + a / (x - h) + b
    y[(y <= lower_limit) | (y >= upper_limit)] = np.nan  # Исключаем значения вне заданных пределов
    return y

# Прямая: y = 15/26x - 23/26
def straight_line(x):
    return (15/26) * x - (23/26)

# Дополнительные точки
additional_points_x = np.linspace(1, 5, 15)
additional_points_y = np.random.uniform(3, 4, 15)

# Генерация данных
x_range = np.linspace(-5, 5, 400)
y_hyperbola = hyperbola(x_range, a=1, b=0, h=0, k=5, upper_limit=10, lower_limit=5)
y_straight_line = straight_line(x_range)

# Создание графика
fig, ax = plt.subplots(figsize=(8, 6))

# График гиперболы
ax.plot(x_range, y_hyperbola, label='Гипербола', linestyle='-', color='purple')

# Добавление дополнительных точек с маркерами
ax.scatter(additional_points_x, additional_points_y, label='Дополнительные точки', color='green', marker='o', facecolors='none')

# График прямой
ax.plot(x_range, y_straight_line, label='Прямая', linestyle='--', color='blue')

# Легенда
ax.legend()

# Заголовок
ax.set_title('Гипербола, Прямая и Дополнительные точки с маркерами')

# Оси координат
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Установка пределов по осям
ax.set_xlim(0, 6)
ax.set_ylim(0, 10)

# Добавление пунктирной сетки и промежуточных делений
ax.grid(True, linestyle='--', which='major')
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())

# Добавление подписей к промежуточным делениям
def minor_tick_formatter(value, pos):
    return f'{value:.2f}'

ax.yaxis.set_minor_formatter(FuncFormatter(minor_tick_formatter))

# Показать график
plt.show()
```
